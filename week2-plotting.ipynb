```python
# Read in data
URL = 'https://raw.githubusercontent.com/allisonhorst/palmerpenguins/main/inst/extdata/penguins.csv'
penguins = pd.read_csv(URL)

penguins.head()
```

Select the bill_length_mm and bill_depth_mm columns in the penguins dataframe and then update the kind parameter to box to make boxplots of the bill length and bill depth.

Create a simple histogram of the flipper length of female gentoo penguins.


```python
penguins_subset = penguins[["bill_length_mm", "bill_depth_mm"]]
# Using plot without subsetting data - a mess again
penguins_subset.plot(kind ='box', title = "Distribution of bill length and depth (mm) among sampled penguins", 
                    xlabel = "Descriptor", ylabel = "Measurement (mm)", color = "hotpink")
```

```python
#Create a simple histogram of the flipper length of female gentoo penguins.
female_gentoos = penguins[(penguins["sex"] == "female") & (penguins["species"]== "Gentoo")]
female_gentoos.head()
```

```python
female_gentoos["flipper_length_mm"].plot(kind = "hist", title = "Distribution of flipper lengths in female gentoo pengiuns", 
                                        xlabel = "Flipper Length (mm)", color = "hotpink")
```
